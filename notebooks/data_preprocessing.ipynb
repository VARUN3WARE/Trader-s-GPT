```python
import pandas as pd

# Load one sample file to inspect (adjust filename as needed)
file_path = '../Output/data_1h/AAPL_1h_data.csv'

# Load with default header
df = pd.read_csv(file_path)
display(df.head(10))
display(df.columns)
```

```python
df = pd.read_csv(file_path, skiprows=2)
display(df.head())
```

```python
display(df.columns)
```

```python
# Rename columns (your CSV may use 'Price' instead of 'Close')
df = df.rename(columns={
    'Price': 'Close',  # if 'Price' means 'Close'
    'Datetime': 'Gmt time',
    'Unnamed: 1': 'Close',
    'Unnamed: 2' : 'High',
    'Unnamed: 3': 'Low',
    'Unnamed: 4' : 'Open',
    'Unnamed: 5' : 'Volume'
})

# Format the datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time']).dt.strftime('%d.%m.%Y %H:%M:%S.000')

# Reorder columns
df = df[['Gmt time', 'Open', 'High', 'Low', 'Close', 'Volume']]
df = df.round({'Open': 5, 'High': 5, 'Low': 5, 'Close': 5, 'Volume': 2})

df.head()
```
